```python
import pandas as pd
import numpy as np
wdf=pd.read_csv("testset.csv")
wdf.columns
```

To make pie charts for different weather conditions in a month. 

```python
cond = wdf['_conds']
```

```python
freq = {} # Apr 1999

for ind,condition in cond.items():
    mo = str(ind.month)
    ye = str(ind.year)
    if ye == '1999' and mo == '4':
        if freq.get(condition) == None:
            freq[condition] = 0
        freq[condition] += 1

y = list(freq.values())
x = list(freq.keys())
plt.pie(y,labels=x)
plt.show()
```
